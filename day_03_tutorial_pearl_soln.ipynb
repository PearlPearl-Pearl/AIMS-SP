```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# NumPy has the functions np.real() and np.imag() functions
# for taking the real and imaginary parts of a number, respectively.
# We can use that as follows:
x = 4.5 - 3j
x_real = np.real(x)
x_imag = np.imag(x)

print("Real part of x : {}; imag part of x: {}".format(x_real, x_imag))
```

```python
# this also applies to complex expontials
z = np.exp(1j * np.pi/4)
z_real = np.real(z)
z_imag = np.imag(z)

print("Real part of z : {}; imag part of z: {}".format(z_real, z_imag))
# is the above correct? use a picture to show it is (or isn't)
```

```python
plt.plot([0, z_real], [0, z_imag], linewidth=2)  # line from origin to point
plt.scatter(z_real, z_imag)              # the point itself
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.grid(True)
plt.axis('scaled')
plt.show()
```

Task 1:  Make an array of complex values according to the following rule:

$v_n = e^{-i2\pi n/N}$, where $n \in [0, N)$. Use different integer N values, like 4, 5, 8 and 9.

Then, plot the $Re(v_n)$ and $Im(v_n)$ as x- and y-values, respectively.

What do you notice about the plots?

What does this tell about the average value of a given set, $<v_n>$?

```python
N = 99

real_v = np.zeros(N)
imag_v = np.zeros(N)

for n in range(N):
    v_n = np.exp(-1j*(2*np.pi*n/N))
    real_v[n] = np.real(v_n)
    imag_v[n] = np.imag(v_n)

plt.scatter(real_v, imag_v)
plt.axis('scaled')
```

Task 2:  Make an array of complex values according to the following rule:

$w_n = e^{-i2\pi m n/N}$, where $n \in [0, N)$. Use different integer N values, like 4, 5, 8 and 9, and various m values like 2, 3, 4. Also try m=0.

Then, plot the $Re(w_n)$ and $Im(w_n)$ as x- and y-values, respectively.

What do you notice about the plots?

What does this tell about the average value of a given set, $<w_n>$, particularly with regards to m?

```python
N = 5
m = 5

real_w = np.zeros(N)
imag_w = np.zeros(N)

for n in range(N):
    w_n = np.exp(-1j*(2*np.pi*n*m)/N)
    real_w[n] = np.real(w_n)
    imag_w[n] = np.real(w_n)


plt.scatter(real_w, imag_w)
plt.axis('scaled')
plt.show()
```

```python
t = np.linspace(-10, 10, 5001)
w_vals = np.linspace(-np.pi, np.pi, 5001)
A_vals =  [1,2,3,4,10]

fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (12, 4))

for idx, A in enumerate(A_vals):
    x = np.array([1 if -(A/2) <= ti <= (A/2) else 0 for ti in t])
    X = A*np.array([np.abs(np.sinc((A*w)/2)) for w in w_vals])
    ax[0].plot(t, x, label=f'A={A}')
    ax[1].plot(t, X, label=f'A = {A}')



ax[0].legend()
ax[1].legend()
ax[0].set_xlabel('t')
ax[1].set_xlabel('$\omega$')
ax[0].set_ylabel('x(t)')
ax[1].set_ylabel('X$(\omega)$')
ax[0].set_ylim([-2,15])
ax[1].set_ylim([-2, 15])
ax[1].grid()
ax[0].grid()

fig.suptitle('Tutorial 2 - Question 3')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

N = 21
n = np.arange(-10, 11)

x = np.zeros(N)
h = np.zeros(N)

x[10] = 1
x[13] = -1
x[14] = 1
x[16] = 2

h[10] = 1
h[11] = -1

y = np.convolve(x, h, mode='same')

fig = plt.figure(figsize=(10,6))

# define grid
gs = fig.add_gridspec(2,2)

ax1 = fig.add_subplot(gs[0,0])   # top left
ax2 = fig.add_subplot(gs[0,1])   # top right
ax3 = fig.add_subplot(gs[1,:])   # bottom spanning both columns

# x plot
ax1.stem(n, x)
ax1.set_title("x[n]")
ax1.set_xticks(n)
ax1.grid()
ax1.axhline(c='r')
ax1.axvline(c='r')

# h plot
ax2.stem(n, h)
ax2.set_title("h[n]")
ax2.set_xticks(n)
ax2.grid()
ax2.axhline(c='r')
ax2.axvline(c='r')

# y plot
ax3.stem(n, y)
ax3.set_title("y[n] = x[n] * h[n]")
ax3.set_xticks(n)
ax3.grid()
ax3.axhline(c='r')
ax3.axvline(c='r')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import convolve

# create a 2D image of a sphere using distance from center
size = 200
x = np.linspace(-1, 1, size)
y = np.linspace(-1, 1, size)
X, Y = np.meshgrid(x, y)

# distance from center
D = np.sqrt(X**2 + Y**2)

# create sphere — inside circle is white, outside is black
image = np.zeros((size, size))
image[D < 0.7] = 255

# edge filter
kernel = np.array([[1, -1],
                   [1, -1]])

# apply edge filter
edge = convolve(image, kernel)

# plot
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(image, cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(edge, cmap='gray')
axes[1].set_title('After Edge Filter')
axes[1].axis('off')

axes[2].imshow(np.abs(edge), cmap='hot')
axes[2].set_title('Edge Magnitude (absolute value)')
axes[2].axis('off')

plt.tight_layout()
plt.show()
```
