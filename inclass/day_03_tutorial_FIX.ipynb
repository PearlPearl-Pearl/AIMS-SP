```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# NumPy has the functions np.real() and np.imag() functions
# for taking the real and imaginary parts of a number, respectively.
# We can use that as follows:
x = 4.5 - 3j
x_real = np.real(x)
x_imag = np.imag(x)

print("Real part of x : {}; imag part of x: {}".format(x_real, x_imag))
```

```python
# this also applies to complex expontials
z = np.exp(1j * np.pi/4)
z_real = np.real(z)
z_imag = np.imag(z)

print("Real part of z : {}; imag part of z: {}".format(z_real, z_imag))
# is the above correct? use a picture to show it is (or isn't)
```
