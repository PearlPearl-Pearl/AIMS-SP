## Question 1

**(a)** Let $a(t) = \delta(t)$. We see that this is a non-periodic function. Thus the Fourier Transform coefficients are given as:

\begin{align*}A(\omega) &= \int_{-\infty}^{\infty} a(t)e^{-i\omega t} dt\\
&= \int_{-\infty}^{\infty} \delta(t)e^{-i\omega t} dt \\
&= e^0\\
&= 1\end{align*}

Thus 
$$A(\omega) \equiv 1\text{ and so } |A(\omega)| \equiv 1$$

Since the Fourier Transform coefficient is purely real, then the phase $\phi(\omega)= 0$

<br><br>

**(b)** $b(t) = \delta(t-3)$

\begin{align*}B(\omega) &= \int_{-\infty}^{\infty} b(t)e^{-i\omega t} dt\\
&= \int_{-\infty}^{\infty} \delta(t-3)e^{-i\omega t} dt\\
&= e^{-i\omega(3)}\\ 
&= \cos(-3\omega) + i\sin(-3\omega)\\
&= \cos(3\omega) - i\sin(3\omega)\end{align*}
From the above, it is easy to conclude that the phase $\phi(\omega) = -3\omega$ and thus we compute the magnitude $|B(\omega)|$ as follows:
$$|B(\omega)| = \sqrt{\cos^2(3\omega) + \sin^2(3\omega)} = \sqrt{1} = 1$$

<br><br>

**(c)** Substituting the values for $A(\omega)$ and $B(\omega)$ into the respective synthesis equations $$a(t) = \frac{1}{2\pi}\int\limits_{-\infty}^{\infty}A(\omega)e^{i\omega t}d\omega \text{ and } b(t) = \frac{1}{2\pi}\int\limits_{-\infty}^{\infty}B(\omega)e^{i\omega t}d\omega$$ we get the following:


\begin{align*}
a(t) &= \frac{1}{2\pi}\int_{-\infty}^{\infty} A(\omega)e^{i\omega t} \, d\omega\\
&= \frac{1}{2\pi}\int_{-\infty}^{\infty} e^{i\omega t} \, d\omega\\
&= \frac{1}{2\pi}\int_{-\infty}^{\infty} \left(\cos(\omega t) + i\sin(\omega t)\right) d\omega\\
&= \frac{1}{2\pi}\left(\int_{-\infty}^{\infty} \cos(\omega t) \, d\omega + i\int_{-\infty}^{\infty} \sin(\omega t) \, d\omega\right)
\end{align*}


Since $\sin(\omega t)$ is an odd function, its integral over a symmetric interval vanishes, leaving:


\begin{align*}
&= \frac{1}{\pi}\int_{0}^{\infty} \cos(\omega t) \, d\omega\end{align*}

This cannot be simplified algebraically any further.


---


\begin{align*}
b(t) &= \frac{1}{2\pi}\int_{-\infty}^{\infty} B(\omega)e^{i\omega t} \, d\omega \\
&= \frac{1}{2\pi}\int_{-\infty}^{\infty} e^{-3i\omega} e^{i\omega t} \, d\omega \\
&= \frac{1}{2\pi}\int_{-\infty}^{\infty} e^{i\omega(t-3)} \, d\omega \\
&= \frac{1}{2\pi}\int_{-\infty}^{\infty} \left(\cos(\omega(t-3)) + i\sin(\omega(t-3))\right) d\omega\\
&= \frac{1}{2\pi}\left(\int_{-\infty}^{\infty} \cos(\omega(t-3)) \, d\omega + i\int_{-\infty}^{\infty} \sin(\omega(t-3)) \, d\omega\right)
\end{align*}


Since $\sin(\omega(t-3))$ is odd in $\omega$, its integral vanishes, leaving:


\begin{align*}
&= \frac{1}{\pi}\int_{0}^{\infty} \cos(\omega(t-3)) \, d\omega
\end{align*}


# What I noticed:
Even though none of the above integrals could be simplified algebraically any further, 
we can make use of a known result, [The Inverse Fourier Transform of 1](https://materia.dfa.unipd.it/salasnich/dfl/dfl.pdf):
\begin{align*}
\mathcal{F}^{-1}\{1\} = \frac{1}{2\pi}\int_{-\infty}^{\infty} e^{i\omega t} \, d\omega = \delta(t)
\end{align*}
Expanding the exponential using Euler's identity and noting that the sine term vanishes, this implies:
\begin{align*}
\frac{1}{2\pi}\int_{-\infty}^{\infty} \cos(\omega t) \, d\omega &= \delta(t)
\end{align*}
and similarly, since $\mathcal{F}^{-1}\{e^{-i\omega t_0}\} = \delta(t - t_0)$, we have:
\begin{align*}
\frac{1}{\pi}\int_{0}^{\infty} \cos(\omega(t-3)) \, d\omega &= \delta(t-3)
\end{align*}
Both synthesis equations therefore correctly recover the original time domain signals $a(t) = \delta(t)$ 
and $b(t) = \delta(t-3)$, confirming the Fourier Transform pair relationship.

# <center><strong>Question 2</strong></center>

Let $f(t) = e^{-a|t|}$. Note that it is a non-periodic function. Thus the Fourier Transform coefficients are given as:
\begin{align*}
F(\omega) &= \int_{-\infty}^{\infty} f(t)e^{-i\omega t} \, dt \\
&= \int_{-\infty}^{\infty} e^{-a|t|}e^{-i\omega t} \, dt \\
&= \int_{-\infty}^{0} e^{at}e^{-i\omega t} \, dt + \int_{0}^{\infty} e^{-at}e^{-i\omega t} \, dt \\
&= \int_{-\infty}^{0} e^{t(a-i\omega)} \, dt + \int_{0}^{\infty} e^{-t(a+i\omega)} \, dt \\
&= \int_{-\infty}^{0} e^{t(i\omega - a)} \cdot \frac{1}{1} \, dt + \int_{0}^{\infty} e^{-t(a+i\omega)} \, dt \\
&= \left[\frac{e^{t(a-i\omega)}}{a - i\omega}\right]_{-\infty}^{0} + \left[\frac{e^{-t(a+i\omega)}}{-(a+i\omega)}\right]_{0}^{\infty}
\end{align*}

We consider the following 2 cases for evaluating the integral at the limits.

### Case 1: a > 0

\begin{align*}
\left[\frac{e^{t(a-i\omega)}}{a - i\omega}\right]_{-\infty}^{0} + \left[\frac{e^{-t(a+i\omega)}}{-(a+i\omega)}\right]_{0}^{\infty} &= \left(\frac{1}{a - i\omega} - 0\right) + \left(0-\left(-\frac{1}{a + i\omega}\right)\right)
\\
&=\frac{1}{a - i\omega} + \frac{1}{a + i\omega} \\[6pt]
&= \frac{(a + i\omega) + (a - i\omega)}{(a - i\omega)(a + i\omega)} \\[6pt]
&= \frac{2a}{a^2 - (i\omega)^2} \\[6pt]
&= \dfrac{2a}{a^2 + \omega^2}
\end{align*}

<br><br>

### Case 2: a $\le$ 0

When $a \leq 0$, consider the first boundary term as $t \to -\infty$:

$$\frac{e^{t(a-i\omega)}}{a - i\omega} \quad \text{as } t \to -\infty$$

The real part of the exponent is $ta$. Since $a \leq 0$ and $t \to -\infty$, we have $ta \to +\infty$, so:

$$e^{t(a-i\omega)} \to \infty$$

The first integral **diverges**. Similarly for the second term, as $t \to +\infty$ the real part of the exponent is $-ta \geq 0$, so $e^{-t(a+i\omega)} \to \infty$ as well.

Therefore, when $a \leq 0$, the Fourier Transform integral does not converge, and $F(\omega)$ **does not exist**.

```python
N = 5001
t = np.linspace(-15, 15, N)
w = np.linspace(-15, 15, N)
s_vals = [0.5,1,2,4]

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,4))

for s in s_vals:
    g = (1/np.sqrt(2*np.pi*(s**2)))*np.exp(-0.5*(t/s)**2)
    G = np.exp(-((w*s)**2)/2)

    ax[0].plot(t, g, label=f'$\sigma$={s}')
    ax[1].plot(w, np.abs(G), label=f'$\sigma$={s}')
    
ax[0].grid()
ax[0].set_xlabel('t')
ax[0].set_ylabel('g(t)')
ax[0].set_title('g(t) against t')
ax[0].set_ylim([-0.3, 1])
ax[0].set_xlim([-10, 10])
ax[0].legend()

ax[1].set_ylim([-0.5, 1.5])
ax[1].set_xlim([-10, 10])
ax[1].set_xlabel('$\omega$')
ax[1].set_ylabel('|G($\omega$)|')
ax[1].set_title('|G($\omega$)| against $\omega$')
ax[1].grid()
ax[1].legend()

fig.suptitle('Assignment 2-Question 3')
plt.tight_layout()
plt.show()
```

Comparing the 2 plots, we see that increasing the value of $\sigma$ in the left plot reduces its height but keeps its support wider, whereas the height of the graphs in the right plot remain unchanged while the support gets narrower. This is consistent with what we studied in class.
