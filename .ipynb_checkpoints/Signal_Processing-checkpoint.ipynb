## <strong><center>Day 01 - Sinusoids</center></strong>

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Plot the following function:
### <center>$y = A \cos{(\omega t - \phi)}$</center>

```python
x = np.linspace(-10, 10, 501)
A = 2.0
w = 1.5
phi = 0

def f(A, w, phi, t):
    return A*(np.cos(w*t - phi))

f1 = f(A, w, phi, x)
# f2 = f(1, 1, 1, x)
```

```python
plt.plot(x, f1)
# plt.plot(x, f2)
plt.show()
```

```python
## lecturer code:
N = 501
t = np.linspace(-10, 10, N) # initialize t
y = np.zeros(N, dtype=float) # initialize y
z = np.zeros(N, dtype=float) # initialize z

A = 2.0
omega = 1.5
phi = 0

for i in range(0, N):
    y[i] = A*np.cos(omega*t[i] - phi)
```

```python
plt.plot(t, y);
# plt.grid()
plt.axhline(c='black')
plt.axvline(c='black')
```

## <strong><center>Day 02 - Sinusoids</center></strong>

```python
A2 = 0.5
omega2 = 3 
phi2 = np.pi/3

for i in range(0, N):
    z[i] = A2*np.cos(omega2*t[i] - phi2)
```

```python
# plt.grid()
plt.axhline(c='black')
plt.axvline(c='black')
plt.plot(t, y)
# plt.plot(z, y)
plt.plot(t, z)
plt.show()
```
